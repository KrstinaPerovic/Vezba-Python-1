```python
sns.histplot(df['Annual Income (k$)'], kde = True)
plt.title("Distribucija godisnjih prihoda")
plt.xlabel("Godisnji prihodi u hiljadama dolara")
plt.ylabel("Frekvencija")
plt.show()
```

Ovaj histogram nam pomaze da razumemo kako se godisnji prihodi raspodeljuju medju kupcima.
Na histogramu vidimo da se godisnji prihodi vecinom krecu od 50-75 hiljada dolara. Takodje, postoje nekoliko kupaca sa visokim prihodina koji se krecu oko 100-130 hiljada dolara, ali su oni redji.
Linija na grafiku nam daje glatku aprokcimaciju distribucije i potvrdjuje da je vecina kupaca u sredisnjem segmentu po pitanju godisnjeg prihoda.

```python
sns.histplot(df['Spending Score (1-100)'], kde = True)
plt.title("Distribucija rejtinga potrosnje")
plt.xlabel("Rejting potrosnje")
plt.ylabel("Frekvencija")
plt.show()
```

Na histogramu mozemo videti da se rejting potrosnje vecinom koncentrise oko srednjih vrednosti, narocito oko 50. Postoje kupci i sa veoma visokim i veoma niskim rejtinom potrosnje, ali oni su redji.
Kao i na prethodnom dijagramu i ovde imamo KDE liniju koja nam daje glatku aprokcimaciju distribucije i pokazuje da vecina kupaca ima srednji rejting potrosnje.


```python
sns.histplot(df['Age'], kde = True, bins = 10)
plt.title("Distribucija starosti")
plt.xlabel("Starost")
plt.ylabel("Frekvencija")
plt.show()
```

Na ovom histogramu vidimo kako se starost kupaca raspodeljuje. Ovim dobijama potpuniju sliku. Vidimo da je akcenat na mladjoj populaciji, posebno oko 30-35 godina.
