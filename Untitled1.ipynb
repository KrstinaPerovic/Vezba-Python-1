Na ovom histogramu vidimo da se godisnji prihod najvise krece od oko 50-75 hiljada dolara. Postoje nekoliko kupaca sa visokim godisnjim prihodom (oko 100-130 hiljada), ali su oni redji.


```python
sns.histplot(df['Spending Score (1-100)'], kde = True)
plt.title("Distribucija rejtinga potrosnje")
plt.xlabel("Rejting potrosnje")
plt.ylabel("Frekvencija")
plt.show()
```

Na histogramu mozemo videti da je najvisi rejting potrosnje oko 50. Postoje kupci i sa veoma niskim i veoma visokim rejtingom potrosnje ali oni su manje brojni.


```python
sns.histplot(df['Age'], bins= 10, kde = True)
plt.title("Distribucija starosti")
plt.xlabel("Starost")
plt.ylabel("Frekvencija")
plt.show()
```

Na osnovu vizuelizacije godina koristika mozemo videti da najveci deo populacije pripada uzrastu od 30 do 50 godina, sa izrazajno grupom mladje publike (20-25 godina)

```python
bins = [10, 20, 30, 40, 50, 60, 70, 80]
labels = ['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-80' ]
df['Age Group'] = pd.cut(df['Age'], bins = bins, labels = labels, right = False)
```

```python
df
```

```python
age_spending_score = df.groupby('Age Group')['Spending Score (1-100)'].mean().reset_index()
```

```python
age_spending_score
```

```python
#Vizuelizacija
sns.barplot(x = 'Age Group', y = 'Spending Score (1-100)', data = age_spending_score)
plt.title("Prosecni Spending Score po starosti")
plt.xlabel("Starost")
plt.ylabel("Sprending Score")
plt.show()
```

Na osnovu grafika zakljucujemo da mladje generacije imaju najvisi Sprening Score, oko 60. Ovo ukazuje da mladje generacije generalno vise trose u trznom centru, sto moze biti znacajna informacija za marketinske aktivnosti.

```python
df.head()
```

```python
gender_spending_score = df.groupby('Gender')['Spending Score (1-100)'].mean().reset_index()

sns.barplot(x = 'Gender', y = 'Spending Score (1-100)', data = gender_spending_score)
plt.title("Prosecni Spending Score po polu")
plt.xlabel("Gender")
plt.ylabel("Sprending_Score")
plt.show()
```

Na osnovu grafika mozemo zakljuciti da zene za nijansu vise trose od muskaraca u datom trznom centru. Vise za nekih 2 do 3 %, ali to u poslovnom smislu moze dosta znaciti.
Prema ovom i prethodnom grafiku neki zakljucak je da kampanja treba biti najvise fokusirana na zenski pol u starosti od 20 do 40 godina.




```python
sns.scatterplot(x = 'Annual Income (k$)', y= 'Spending Score (1-100)', data= df)
plt.title("Odnos izmedju godisnjeg prihoda i potrosnje moci")
plt.xlabel("Godisnji prihod (u hiljadama dolara)")
plt.ylabel("Sprending_Score")
```

Svaka tacka predstavlja jednog korisnika, sto omogucava uvid u obrasce ponasanja. Rezultati pokazuju da najveci broj kupaca spada u srednji raspon prihoda (40.000-60.000) i da im je umerena potrosnja, sto predstavlja najstabilniji deo trzista. Takodje, imamo i dva ekstrema: korisnici sa niskim primanjima koji ipak imaju visok spending score, kao i oni sa visokim primanjima koji malo trose.
Na osnovu grafika zakljucujemo da korelacija izmedju prihoda i potrosnje nije dovoljno jaka da bi bila glavni kriterijum za segmentaciju kupaca.
